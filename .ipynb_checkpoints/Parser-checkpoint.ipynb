```python
import re, json, csv
import pandas as pd
SOURCE_FILE = "data/output.txt"
OUTPUT_FILE = "data/posts_data.csv"
```

```python
# open text file
with open(SOURCE_FILE, encoding="utf-8-sig") as file:
    contents = file.read()
    print(contents)
```

```python
# split file into posts
posts = contents.split("\nVM439:26 --GGCB--\nVM439:16 name\nVM439:25 ")
print(posts)
```

```python
# get title, body, and submission details from each post
posts_data = []

with open(OUTPUT_FILE, 'w', encoding='utf-8-sig') as outfile:
    wr = csv.writer(outfile, quoting=csv.QUOTE_ALL)  # Make writer
    wr.writerow(['Title', 'Body', 'Submitted'])  # Writer colum names
    
    for post in posts:
        title = post.splitlines()[0][1:].replace('ADMUFreedomWall', '')
        body = re.search('(?s)(.*)Submitted:', post)

        if body:
            body = body.group(1).split('\n',2)[2].strip()

        submitted = re.search('(?s)Submitted:(.*)', post)

        if submitted:
            submitted = submitted.group(1).strip()
            
        post_contents = [title, body, submitted]
        posts_data.append(post_contents)
        
        wr.writerow(post_contents)  # Write row to csv
```

```python
# convert list to df for checking
posts_df = pd.DataFrame(data=posts_data, columns=['Title', 'Body', 'Submitted'])
posts_df.head()
```
